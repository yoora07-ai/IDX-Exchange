```python
# import libraries
import pandas as pd
import glob as g
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
```

```python
# Import files and sort them 
csv_files = sorted(g.glob("./*.csv"))
```

```python
# Check row counts, and column counts in each csv file 
summary = []

for file in csv_files:
    df = pd.read_csv(file)
    summary.append({
        'file': file,
        'num_rows': df.shape[0],
        'num_columns': df.shape[1]
    })

summary_df = pd.DataFrame(summary)
summary_df
```

```python
# Check row counts in each csv filetotal number of rows in all csv files
total = 0 
for f in csv_files:
    df = pd.read_csv(f)
    rows = len(df)
    total += rows

print(f"\nTotal number of rows(Before appending): {total:,}rows")
```

```python
# Appending csv files
df_merge = pd.concat(
    [pd.read_csv(f) for f in csv_files],
    ignore_index= True
)

# Export csv file
df_merge.to_csv("CRMLSSold_merge.csv", index= False)
```

```python
# check row number after appending csv files
print(f"Total number of rows(After appending): {len(df_merge):,}rows")
```

```python
# vaildation

# Row check
if total == len(df_merge):
    print("Row count matches")
else:
    print("Row count mismatch")

# Column check
if df_merge.shape[1] == summary_df['num_columns'].max():
    print("Column count matches expected")
else:
    print("Column count mismatch")
```

```python
# Compare column names across files
all_columns = set()
for f in csv_files:
    df = pd.read_csv(f, nrows=0)  # Read only header
    all_columns.update(df.columns)

print(f"Total unique columns: {len(all_columns)}")
print(f"df_merge column count:  {df_merge.shape[1]}")

# Check which files are missing columns
for f in csv_files:
    df = pd.read_csv(f, nrows=0)
    missing = all_columns - set(df.columns)
    if missing:
        print(f"{f}: Missing columns --> {missing}")
```
