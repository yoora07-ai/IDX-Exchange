## 2. Missing data check(null)

```python
# null count
null_count = listed.isnull().sum()

# null percentage (%)
null_percent = (listed.isnull().sum() / len(listed)) * 100

# data frame
null_summary = pd.DataFrame({
    'null_count': null_count,
    'null_percent': null_percent
}).sort_values(by='null_percent', ascending=False)

null_summary['null_percent'] = null_summary['null_percent'].round(2)
null_summary
```

## 3. Check duplicates

### 3.1 duplicated columns

```python
# check duplicate columns
pairs = []

for col in listed.columns:
    if '.' in col:
        base_col = col.split('.')[0]
        if base_col in listed.columns:
            pairs.append((base_col, col))

pairs
```

```python
# Check if duplicate column pairs have identical values row-by-row
for base, dup in pairs:
    same_ratio = (listed[base] == listed[dup]).mean()
    print(f"{base} vs {dup}: {same_ratio:.2f}")
```
