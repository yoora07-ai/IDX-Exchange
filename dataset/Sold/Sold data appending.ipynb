```python
import pandas as pd
import glob as g
```

```python
# Import files and sort them 
csv_files = sorted(g.glob("./*.csv"))
```

```python
# Check row counts in each csv filetotal number of rows in all csv files
total = 0 
for f in csv_files:
    df = pd.read_csv(f)
    rows = len(df)
    total += rows
    print(f"{f}: {rows:,}rows")

print(f"\nTotal number of rows(Before appending): {total:,}rows")
```

```python
# Appending csv files
df_merge = pd.concat(
    [pd.read_csv(f) for f in csv_files],
    ignore_index= True
)

# Export csv file
df_merge.to_csv("CRMLSold_merge.csv", index= False)
```

```python
# check row number after appending csv files
print(f"Total number of rows(After appending): {len(df_merge):,}rows")
```

```python
# vaildation
if(total == len(df_merge)):
    print("the total row # before and after appending is same (Appending success)")
else:
    print("the total row # before and after appending is different (Appending failure)")
```
