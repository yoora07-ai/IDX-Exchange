### Q4. What percentage of homes sold above vs. below list price?

```python
#Sold Above vs. Below List Price ───────────────────────────────────────
# ⚠️ Listing data only — ClosePrice exists for ~26% of records (closed listings only)
# This analysis is limited to that subset
sold = listed.dropna(subset=['ClosePrice', 'ListPrice']).copy()
sold['price_ratio'] = sold['ClosePrice'] / sold['ListPrice']

above = (sold['price_ratio'] > 1).sum()
at    = (sold['price_ratio'] == 1).sum()
below = (sold['price_ratio'] < 1).sum()
total = len(sold)

print("\n=== Sold Price vs. List Price (closed listings only, n={:,}) ===".format(total))
print(f"Above list price:  {above:,} ({above/total*100:.1f}%)")
print(f"At list price:     {at:,}  ({at/total*100:.1f}%)")
print(f"Below list price:  {below:,} ({below/total*100:.1f}%)")
print(f"\nMedian Close/List ratio: {sold['price_ratio'].median():.4f}")
print(f"Mean Close/List ratio:   {sold['price_ratio'].mean():.4f}")

```

### Q5. Are there any apparent date consistency issues (e.g., close date before listing date)?

```python
# Date Consistency Issues ───────
listed['CloseDate_dt']    = pd.to_datetime(listed['CloseDate'], errors='coerce')
listed['ListingDate_dt']  = pd.to_datetime(listed['ListingContractDate'], errors='coerce')
listed['PurchaseDate_dt'] = pd.to_datetime(listed['PurchaseContractDate'], errors='coerce')

# Close before listing
c_before_l = listed.dropna(subset=['CloseDate_dt', 'ListingDate_dt'])
close_before_list = (c_before_l['CloseDate_dt'] < c_before_l['ListingDate_dt']).sum()

# Purchase before listing
p_before_l = listed.dropna(subset=['PurchaseDate_dt', 'ListingDate_dt'])
purchase_before_list = (p_before_l['PurchaseDate_dt'] < p_before_l['ListingDate_dt']).sum()

# Close before purchase
c_before_p = listed.dropna(subset=['CloseDate_dt', 'PurchaseDate_dt'])
close_before_purchase = (c_before_p['CloseDate_dt'] < c_before_p['PurchaseDate_dt']).sum()

print("\n=== Date Consistency Issues ===")
print(f"Close date before listing date:    {close_before_list}")
print(f"Purchase date before listing date: {purchase_before_list}")
print(f"Close date before purchase date:   {close_before_purchase}")

```

### Q6.Which counties have the highest median prices?

```python
# Median Price by County
# ListPrice — all records
county_list = (
    listed.groupby('CountyOrParish')['ListPrice']
    .agg(median_list_price='median', total_listings='count')
    .sort_values('median_list_price', ascending=False)
    .reset_index()
)
print("\n=== Median List Price by County (Top 15, all listings) ===")
print(county_list.head(15).to_string(index=False))

# ClosePrice — closed listings only
# Limited to ~26% of records
county_close = (
    listed.dropna(subset=['ClosePrice'])
    .groupby('CountyOrParish')['ClosePrice']
    .agg(median_close_price='median', sold_count='count')
    .sort_values('median_close_price', ascending=False)
    .reset_index()
)
print("\n=== Median Close Price by County (Top 15, closed only) ===")
print(county_close.head(15).to_string(index=False))
```
