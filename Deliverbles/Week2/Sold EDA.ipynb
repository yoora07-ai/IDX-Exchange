### Q3. What is the price ratio (ClosePrice / ListPrice)?

```python

sold = df.dropna(subset=['ClosePrice', 'ListPrice']).copy()
sold['price_ratio'] = sold['ClosePrice'] / sold['ListPrice']

above = (sold['price_ratio'] > 1).sum()
at    = (sold['price_ratio'] == 1).sum()
below = (sold['price_ratio'] < 1).sum()
total = len(sold)

print(f"Above list price: {above:,} ({above/total*100:.1f}%)")
print(f"At list price:    {at:,} ({at/total*100:.1f}%)")
print(f"Below list price: {below:,} ({below/total*100:.1f}%)")
print(f"Median ratio:     {sold['price_ratio'].median():.4f}")
```

### Q4. What percentage of homes sold above vs. below list price?

```python
# ── Q4. Sold Above vs. Below List Price ─────────────────────────────────────
sold = df.dropna(subset=['ClosePrice', 'ListPrice']).copy()
sold['price_ratio'] = sold['ClosePrice'] / sold['ListPrice']

above = (sold['price_ratio'] > 1).sum()
at    = (sold['price_ratio'] == 1).sum()
below = (sold['price_ratio'] < 1).sum()
total = len(sold)

print("=== Q4. Sold Price vs. List Price ===")
print(f"Total closed listings:  {total:,}")
print(f"Above list price:       {above:,} ({above/total*100:.1f}%)")
print(f"At list price:          {at:,} ({at/total*100:.1f}%)")
print(f"Below list price:       {below:,} ({below/total*100:.1f}%)")
print(f"Median Close/List ratio:{sold['price_ratio'].median():.4f}")
print(f"Mean Close/List ratio:  {sold['price_ratio'].mean():.4f}")
```

### Q5. Are there any apparent date consistency issues (e.g., close date before listing date)?

```python
# ── Q5. Date Consistency Issues ─────────────────────────────────────────────
df['CloseDate_dt']    = pd.to_datetime(df['CloseDate'], errors='coerce')
df['ListingDate_dt']  = pd.to_datetime(df['ListingContractDate'], errors='coerce')
df['PurchaseDate_dt'] = pd.to_datetime(df['PurchaseContractDate'], errors='coerce')

# Close before listing
c_before_l = df.dropna(subset=['CloseDate_dt', 'ListingDate_dt'])
close_before_list = (c_before_l['CloseDate_dt'] < c_before_l['ListingDate_dt']).sum()

# Purchase before listing
p_before_l = df.dropna(subset=['PurchaseDate_dt', 'ListingDate_dt'])
purchase_before_list = (p_before_l['PurchaseDate_dt'] < p_before_l['ListingDate_dt']).sum()

# Close before purchase
c_before_p = df.dropna(subset=['CloseDate_dt', 'PurchaseDate_dt'])
close_before_purchase = (c_before_p['CloseDate_dt'] < c_before_p['PurchaseDate_dt']).sum()

print("=== Q5. Date Consistency Issues ===")
print(f"Close before listing date:    {close_before_list:,}")
print(f"Purchase before listing date: {purchase_before_list:,}")
print(f"Close before purchase date:   {close_before_purchase:,}")
```


### Q6.Which counties have the highest median prices?

```python
# ── Q6. County Median Prices ─────────────────────────────────────────────────
county = (
    df.dropna(subset=['ClosePrice'])
    .groupby('CountyOrParish')
    .agg(
        median_close_price=('ClosePrice', 'median'),
        median_list_price=('ListPrice', 'median'),
        sold_count=('ClosePrice', 'count')
    )
    .sort_values('median_close_price', ascending=False)
    .reset_index()
)

county['median_close_price'] = county['median_close_price'].apply(lambda x: f"${x:,.0f}")
county['median_list_price']  = county['median_list_price'].apply(lambda x: f"${x:,.0f}")

print("=== Q6. Median Close Price by County (Top 15) ===")
print(county.head(15).to_string(index=False))
```
